```python
import os 


import pandas as pd

```

```python
# from root to get all file path 

file_path = []  # saved all file path 

def file_path_generator():
    root = 'OpinRank/cars/data/'

    for year in os.listdir(root):
        if year.endswith('csv'):
            continue   # skip .csv file
        for file in os.listdir(root+year):
            file_path.append(root+year+'/'+file)
            
file_path_generator()
```

```python
makes = []
models = []
years = []
date_list = []
authors = []
reviews = []
favorites = []


for file in file_path:
#     counter = 0
    temp_date = []
    temp_autor = []
    temp_review = []
    temp_favorite = []
    with open(file=file, mode='r', encoding='ISO-8859-1') as f:
        while True:
            line = f.readline()

            if line.startswith('<DATE>'):
                temp_date.append(line.replace('<DATE>', '').replace('</DATE>', ''))
            if line.startswith('<AUTHOR>'):
                temp_autor.append(line.replace('<AUTHOR>', '').replace('</AUTHOR>', ''))
            if line.startswith('<TEXT>'):
                temp_review.append(line.replace('<TEXT>', '').replace('</TEXT>', ''))
            if line.startswith('<FAVORITE>'):
                temp_favorite.append(line.replace('<FAVORITE>', '').replace('</FAVORITE>', ''))
#             if line.startswith('<DOC>'):
#                 counter += 1 
            if not line:
                break         # break when file ends

        make = file.split('/')[-1].split('_')[1]
        model = file.split('/')[-1].split('_')[-1]
        year = file.split('/')[-1].split('_')[0]

        makes += [make for _ in range(len(temp_review))]
        models += [model for _ in range(len(temp_review))]
        years += [year for _ in range(len(temp_review))]
        date_list += temp_date
        authors += temp_autor
        reviews += temp_review
        favorites += temp_favorite
        
print(len(makes))
print(len(models))
print(len(years))
print(len(date_list))
print(len(authors))
print(len(reviews))
print(len(favorites))

df = pd.DataFrame(columns=['Make', 'Model', 'Year', 'Author', 'Date', 'Reviews', 'Favorite'])
df['Make'] = makes
df['Model'] = models
df['Year'] = years
df['Author'] = authors
df['Date'] = date_list
df['Reviews'] = reviews
df['Favorite'] = favorites

df.to_csv('init_table.csv', index=False)
    
    
# detail_extractor()
# csv_generator()
```
